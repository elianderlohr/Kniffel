```python
def send_step(dice, env, action, score):
    print(f"State old: {env.kniffel.get_state()}")
    action_type = "roll_dice" if action >= 13 and action <= 44 else "finish"
    print(f"Action ({action_type}): {action}")
    if len(dice) > 0:
        env.mock(dice)
    n_state, reward, done, info = env.step(action)
    print(f"Reward: {reward}")
    print(f"Score: {score + reward}")
    print(f"Done: {done}")
    print(f"State new: {n_state}")
    print(f"Round left: {env.kniffel.turns_left()}")
    print(f"Info: {info}")
    if done:
        print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
    print()
    print()
    
    return reward
```
